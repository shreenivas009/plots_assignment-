```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(0) 
n = 1000 
feature_1 = np.random.normal(0, 1, n)
feature_2 = 0.8 * feature_1 + np.random.normal(0, 0.2, n)  
feature_3 = 0.5 * feature_1 + np.random.normal(0, 0.5, n)  
feature_4 = np.random.normal(0, 1, n)  

data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

sales_pivot = df_sales.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(sales_pivot, annot=True, cmap='YlGnBu', linewidths=0.5)

plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

plt.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function.

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('3D Surface Plot of z = sin(√(x² + y²))')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5, color='teal')

for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=12, ha='right')

plt.title('Bubble Chart of Population vs GDP by Country')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

plt.show()
```

```python
''' BOKEH ASSIGNMENT '''
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

p.line(x, y, legend_label="sin(x)", line_width=2, color='navy')

output_notebook() 
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

np.random.seed(42)
n_points = 100
x = np.random.random(size=n_points) * 100  
y = np.random.random(size=n_points) * 100  
sizes = np.random.random(size=n_points) * 20 + 10  
colors = np.random.random(size=n_points) * 100  

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))
p = figure(title="Bokeh Scatter Plot", x_axis_label='x', y_axis_label='y')

p.scatter('x', 'y', size='sizes', color=color_mapper, source=source, fill_alpha=0.6)

output_notebook()  
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts',
           height=400, width=600, toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

output_notebook()  
show(p)
```
